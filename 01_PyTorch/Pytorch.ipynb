# BTWN 01

Cho y = 5x^5 + 6x^4 - 3x + 1

Cho biết độ dốc của đa thức trên ở điểm nào

```python
import torch

# 1. Định nghĩa điểm cần tính đạo hàm (ví dụ x = 1.0)
# requires_grad=True là bắt buộc để PyTorch theo dõi các phép toán và tính đạo hàm
x = torch.tensor([1.0], requires_grad=True)

# 2. Định nghĩa hàm số y = 5x^5 + 6x^4 - 3x + 1
y = 5 * x**5 + 6 * x**4 - 3 * x + 1

# 3. Tính đạo hàm (Backpropagation)
y.backward()

# 4. Lấy giá trị độ dốc tại điểm x
slope = x.grad.item()

print(f"Độ dốc tại x = {x.item()} là: {slope}")
```

# BTVN 02

Tạo một tensor ban đầu có giá trị là 2

Định nghĩa hàm số và tính gradient

y = x^3 + 2x^2 + 5x + 1

Hãy tính dy/dx tại giá trị của x

Dùng phương pháp Gradient Descent với learning rate alpha = 0.1 để cập nhật giá trị x trong quá trình lặp

```python
import torch

# Khởi tạo x
x = torch.tensor([2.0], requires_grad=True)
learning_rate = 0.01

print(f"Giá trị ban đầu x: {x.item()}")
print("-" * 30)

# Vòng lặp 50 lần
for i in range(50):
    # Tính toán hàm số và đạo hàm
    y = x**3 + 2*x**2 + 5*x + 1
    y.backward()
    
    # Lưu giá trị gradient hiện tại để in
    current_grad = x.grad.item()
    
    # Cập nhật x
    with torch.no_grad():
        x -= learning_rate * x.grad
        x.grad.zero_() # Reset gradient
        
    # In kết quả mỗi 3 vòng lặp để theo dõi
    if (i + 1) % 3 == 0 or i == 0:
        print(f"Vòng {i+1}: x = {x.item():.4f}, grad = {current_grad:.4f}")

print("-" * 30)
print(f"Giá trị cuối cùng của x: {x.item()}")
```

## BTVN 03

Tạo một tập dữ liệu giả lập với:
	•	x là số giờ học (ngẫu nhiên từ 1 – 10)
	•	y là số điểm theo công thức:

y = 3x + 5 + noise

Trong đó noise là một giá trị ngẫu nhiên nhỏ

### Yêu cầu:
	1.	Khởi tạo tham số w và b ngẫu nhiên với requires_grad = True
	2.	Tính MSE
	3.	Tính Gradient
	4.	Cập nhật tham số w và b bằng Gradient Descent với learning rate alpha = 0.01
	5.	Lặp lại quá trình trên trong 100 vòng lặp để xem sự hội tụ của tham số

```python
import torch

# Tạo tập dữ liệu giả lập 
torch.manual_seed(42) # Cố định seed để kết quả giống nhau mỗi lần chạy

# Tạo 20 mẫu dữ liệu
# X có giá trị ngẫu nhiên từ 1 đến 10
X = torch.rand(20) * 9 + 1  
noise = torch.randn(20) * 2  # Nhiễu ngẫu nhiên (standard normal distribution)

# Y theo công thức y = 3x + 5 + noise
Y_thuc_te = 3 * X + 5 + noise

print(f"Dữ liệu mẫu (5 điểm đầu):")
for i in range(5):
    print(f"  x: {X[i]:.2f} -> y: {Y_thuc_te[i]:.2f}")
print("-" * 50)

# Khởi tạo tham số w và b
w = torch.tensor([3.0], requires_grad=True) # Giả sử bắt đầu từ 0
b = torch.tensor([5.0], requires_grad=True)

learning_rate = 0.01

print(f"Tham số khởi tạo: w = {w.item():.2f}, b = {b.item():.2f}")
print("-" * 50)

# Vòng lặp huấn luyện (100 lần)
for epoch in range(100):
    
    # 1. Tính toán giá trị dự đoán (Forward pass)
    Y_du_doan = w * X + b
    
    # 2. Tính MSE 
    # Công thức: trung bình của (y_mũ - y_thực)^2
    loss = (Y_du_doan - Y_thuc_te).pow(2).mean()
    
    # 3. Tính Gradient (Backward pass)
    loss.backward()
    
    # 4. Cập nhật tham số
    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad
        
        # Reset gradient về 0 sau khi cập nhật
        w.grad.zero_()
        b.grad.zero_()
        
    # In thông tin mỗi 10 vòng lặp
    if (epoch + 1) % 10 == 0:
        print(f"Vòng {epoch+1:3d}: Loss = {loss.item():.4f} | w = {w.item():.4f}, b = {b.item():.4f}")

print("-" * 50)
print("KẾT QUẢ CUỐI CÙNG:")
print(f"Mô hình học được:     w = {w.item():.4f}, b = {b.item():.4f}")
```

## BTVN 05
### Tạo tensor với
 - empty
 - zeros
 - ones
 - random
 - reshape với view và view as

```python
import torch

print("TẠO TENSOR")

# 1. Empty: Tạo tensor chưa khởi tạo
# Lưu ý: Giá trị sẽ là "rác" (bất kỳ giá trị nào đang có trong ô nhớ đó), không phải luôn là 0.
t_empty = torch.empty(2, 3)
print(f"1. Empty (2x3):\n{t_empty}\n")

# 2. Zeros: Tạo tensor toàn số 0
t_zeros = torch.zeros(2, 3)
print(f"2. Zeros (2x3):\n{t_zeros}\n")

# 3. Ones: Tạo tensor toàn số 1
t_ones = torch.ones(2, 3)
print(f"3. Ones (2x3):\n{t_ones}\n")

# 4. Random: Tạo tensor ngẫu nhiên
# torch.rand tạo số ngẫu nhiên trong khoảng [0, 1) (phân phối đều)
t_rand = torch.rand(2, 3)
print(f"4. Random (2x3):\n{t_rand}\n")


print("RESHAPE VỚI VIEW")

# Tạo một tensor gốc có 12 phần tử (từ 0 đến 11)
t_goc = torch.arange(12) 
print(f"Tensor gốc (1D - 12 phần tử):\n{t_goc}\n")

# 5. Sử dụng .view()
# Biến đổi thành ma trận 3 hàng, 4 cột
# Lưu ý: Tích các chiều (3 * 4) phải bằng tổng số phần tử (12)
t_view = t_goc.view(3, 4)
print(f"5. View (3x4):\n{t_view}\n")

# Mẹo: Dùng -1 để PyTorch tự tính chiều còn lại
# Ví dụ: Muốn 2 hàng, còn bao nhiêu cột thì tự tính --> view(2, -1)
t_view_auto = t_goc.view(2, -1)
print(f"5. View tự động (2 hàng, tự tính cột):\n{t_view_auto}\n")


print("RESHAPE VỚI VIEW_AS")

# 6. Sử dụng .view_as()
# Giả sử ta có một tensor đích (target) nào đó
t_mau = torch.zeros(4, 3) # Tensor mẫu kích thước 4x3

# Ta muốn t_goc có hình dạng GIỐNG NHƯ t_mau
t_view_as = t_goc.view_as(t_mau)

print(f"Tensor mẫu shape: {t_mau.shape}")
print(f"6. View_as (biến t_goc theo t_mau):\n{t_view_as}")
```
