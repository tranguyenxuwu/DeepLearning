```python
import pandas as pd
import numpy as np
```

```python
# Đọc dữ liệu
df = pd.read_csv('Life Expectancy Data.csv')

# Xử lý tên cột: loại bỏ khoảng trắng thừa ở đầu/cuối để dễ thao tác
df.columns = df.columns.str.strip()
```

```python
# 1. In ra 2 dòng đầu tiên và 2 dòng cuối cùng của DataFrame.
print("2 dòng đầu tiên:")
print(df.head(2))
print("\n2 dòng cuối cùng:")
print(df.tail(2))
```

```python
# 2. In ra kích thước (shape) của DataFrame.
print("Kích thước (shape):", df.shape)
```

```python
# 3. In ra tên các đặc trưng (các cột) của DataFrame.
print("Các cột:", df.columns.tolist())
```

```python
# 4. In ra bảng thống kê mô tả bằng hàm .describe().
print(df.describe())
```

```python
# 5. Xóa cột Hepatitis B và Population
df.drop(columns=['Hepatitis B', 'Population'], inplace=True)
print("Đã xóa cột Hepatitis B và Population")
```

```python
# 6. Chuyển đổi cột Status sang dạng số: 0 cho Developing, 1 cho Developed
df['Status'] = df['Status'].map({'Developing': 0, 'Developed': 1})
print("Đã chuyển đổi cột Status. Các giá trị unique còn lại:", df['Status'].unique())
```

```python
# 7. Đổi tên cột thinness 1-19 years thành thinness 10-19 years.
for col in df.columns:
    if 'thinness' in col and '1-19' in col:
        print(f"Đổi tên cột '{col}' -> 'thinness 10-19 years'")
        df.rename(columns={col: 'thinness 10-19 years'}, inplace=True)
```

```python
# 8. Lấy tất cả các cột ngoại trừ Life Expectancy, chuyển sang mảng NumPy và lưu vào biến X.
if 'Life expectancy' in df.columns:
    X = df.drop(columns=['Life expectancy']).values
    print("Shape of X:", X.shape)
else:
    print("Không tìm thấy cột 'Life expectancy'")
```

```python
# 9. Lấy cột Life Expectancy, chuyển sang mảng NumPy và lưu vào biến y.
if 'Life expectancy' in df.columns:
    y = df['Life expectancy'].values
    print("Shape of y:", y.shape)
    print("5 giá trị đầu của y:", y[:5])
```

## Bài Tập 2
1. Kiểm tra xem mỗi cột có bao nhiêu giá trị bị thiếu (missing data / NaN).
2. Xử lý tất cả dữ liệu bị thiếu bằng cách thay thế bằng giá trị trung bình (mean) của từng cột tương ứng.
3. Thực hiện groupby theo quốc gia (Country).
    - Quốc gia nào có tuổi thọ trung bình thấp nhất?
    - Quốc gia nào có tuổi thọ trung bình cao nhất?
4. Thực hiện groupby theo Status (Developed / Developing).
    - Có sự khác biệt rõ rệt nào về tuổi thọ trung bình giữa các quốc gia phát triển và đang phát triển hay không?
5. Tạo một DataFrame mới bằng tay gồm 2 cột:
    - Cột thứ nhất là ID, có giá trị giống với cột Country
    - Cột thứ hai là Noise_level, điền giá trị ngẫu nhiên (tùy ý)
6. Gộp (merge) hai DataFrame lại với nhau dựa trên cột ID.

```python
# 1. Kiểm tra xem mỗi cột có bao nhiêu giá trị bị thiếu
print("Số lượng giá trị thiếu trên mỗi cột:")
print(df.isnull().sum())
```

```python
# 2. Xử lý tất cả dữ liệu bị thiếu bằng cách thay thế bằng giá trị trung bình (mean) của từng cột tương ứng.
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

print("Đã thay thế NaN bằng mean. Kiểm tra lại số lượng NaN:")
print(df.isnull().sum())
```

```python
# 3. Thực hiện groupby theo quốc gia (Country).
country_life_expectancy = df.groupby('Country')['Life expectancy'].mean()

min_country = country_life_expectancy.idxmin()
min_val = country_life_expectancy.min()

max_country = country_life_expectancy.idxmax()
max_val = country_life_expectancy.max()

print(f"Quốc gia có tuổi thọ thấp nhất: {min_country} ({min_val:.2f} năm)")
print(f"Quốc gia có tuổi thọ cao nhất: {max_country} ({max_val:.2f} năm)")
```

```python
# 4. Thực hiện groupby theo Status (Developed / Developing).
status_life_expectancy = df.groupby('Status')['Life expectancy'].mean()
print("Tuổi thọ trung bình theo Status (0: Developing, 1: Developed):")
print(status_life_expectancy)

diff = status_life_expectancy[1] - status_life_expectancy[0]
print(f"\nChênh lệch tuổi thọ giữa Developed và Developing: {diff:.2f} năm")
```

```python
# 5. Tạo một DataFrame mới bằng tay gồm 2 cột:

# Lấy danh sách quốc gia duy nhất
countries = df['Country'].unique()

# Tạo DataFrame mới
noise_df = pd.DataFrame({
    'ID': countries,
    'Noise_level': np.random.rand(len(countries))
})

print("DataFrame noise_df mới tạo (5 dòng đầu):")
print(noise_df.head())
```

```python
# 6. Gộp (merge) hai DataFrame lại với nhau dựa trên cột ID.
merged_df = pd.merge(df, noise_df, left_on='Country', right_on='ID', how='left')

print("Kích thước sau khi merge:", merged_df.shape)
print("5 dòng đầu của merged_df:")
print(merged_df[['Country', 'ID', 'Noise_level']].head())
```
